```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load crops data
cd = pd.read_csv('../Data-raw/sdata.csv')
cd.head()
```

```python
cd.tail()
```

```python
cd.describe()
```

```python
tempData = pd.DataFrame(columns = ['N','P','K', 'pH', 'temperature', 'rainfall', 'crop'])
```

```python
for i in range(0, cd.shape[0]):
    crop = cd.crop.iloc[i]
#     print(crop)
# format(myFloat, '.2f')
    N = (cd['nitrogen min'].iloc[i] + cd['nitrogen max'].iloc[i]) / 2
    K = (cd['potasium min'].iloc[i] + cd['potasium max'].iloc[i]) / 2
    P = (cd['phosphorus min'].iloc[i] + cd['phosphorus max'].iloc[i]) / 2
    pH = (cd['pH min'].iloc[i] + cd['pH max'].iloc[i]) / 2
    temperature = (cd['temperature min'].iloc[i] + cd['temperature max'].iloc[i]) / 2
    rainfall = (cd['precipitation min'].iloc[i] + cd['precipitation max'].iloc[i]) / 2
    d = {"N": N, "P":P, "K": K, "pH": pH, "temperature": temperature, "rainfall": rainfall, 'crop': crop}
#     print(d)
    tempData = tempData.append(d,ignore_index = True)

# tempData.head()
```

```python
tempData.head()
```

```python
tempData.to_csv("../Data-processed/s_data_processed.csv",index=False)
```

```python
processed_data = pd.read_csv('../Data-processed/s_data_processed.csv')
```

```python
# Function for lowering the cases
def change_case(i):
    i = i.replace(" ", "")
    i = i.lower()
    return i
```

```python
processed_data.head()
```

```python
processed_data['crop'] = processed_data['crop'].apply(change_case)
```
